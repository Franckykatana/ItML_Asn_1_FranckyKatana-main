This is an example of how we can use an external EDA class to automate some of the data exploration steps.

```python
df = pd.read_csv("heart.csv")
df.head()
```

Explanation:

df.info() provides a summary of our DataFrame. This includes the number of non-null entries in each column, the data type of each column, and the memory usage.


```python
df.info()
```

### Load External EDA Class

We initialize an instance of the edaDF class
we pass the DataFrame df and the name of the target variable "HeartDisease" to the constructor
And we print the variable

```python
df_eda = ml_utils.edaDF(df,"HeartDisease")
print(df_eda.giveTarget())
```

### Provide Configuration to the EDA Class

Explanation:

We setCat and setNum methods to define which columns in our dataset are categorical and which are numerical.

```python
df_eda.setCat(["Sex", "ChestPainType", "RestingECG", "ExerciseAngina", "ST_Slope"])
df_eda.setNum(["Age", "RestingBP", "Cholesterol", "FastingBS", "MaxHR", "Oldpeak"])
```

### Perform EDA

Explanation:
<ul>
<li>The df_eda.countPlots method generates count plots for each categorical variable in our dataset.
<li>The splitTarg=True splits the plots by the target variable to see how the categories distribute across different classes of the target variable.
<li>The df_eda.histPlots is use to create a histogram
<li>kde=True adds a Kernel Density to smooth the histogram.
<li>splitTarg=True splits the histograms by the target variable.
</ul>

```python
df_eda.countPlots(splitTarg=True)
df_eda.histPlots(kde=True, splitTarg=True)
```

```python
df_eda.fullEDA
```

Explanation:

This method compiles the entire EDA process into an interactive tabbed layout.
It includes data information, count plots for categorical data, and histogram plots for numerical data, all accessible through different tabs.
